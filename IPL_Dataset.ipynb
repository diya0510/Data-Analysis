**IPL DataSet Analysis**

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
ipl=pd.read_csv('ipl.csv')
ipl.head()
```

```python
#number of matches played and the number of attributes in the given dataset
ipl.shape
```

```python
#the player with maximum number of player of the match titles
ipl['player_of_match'].value_counts()[0:5]
```

**BARPLOT**

```python
#barplot representing the times a player has won the title of player_of_match
#creating two lists to be passed in the bar method of matplotlib
a=list(ipl['player_of_match'].value_counts()[0:5].keys())
b=list(ipl['player_of_match'].value_counts()[0:5])
#making the bar plot
plt.bar(a,b)
plt.xlabel("Players")
plt.ylabel("Frequency")
plt.title('Player of The Match')
plt.show()
```

```python
#frequency of result of different categories
ipl['result'].value_counts()
```

```python
#toss win with respect to each team
ipl['toss_winner'].value_counts()
```

```python
#top 5 teams in winning toss
ipl['toss_winner'].value_counts()[0:5]
```

```python
#teams which won the match by batting first
winbr=ipl[ipl['win_by_runs']!=0]
winbr
```

```python
#first 5 teams which won by batting first
winbr.head()
```

```python
#histogram visualizatio of the temas which won by batting first
plt.hist(winbr['win_by_runs'],color='green')
plt.xlabel('Runs')
plt.ylabel('No_of_Matches')
plt.show
```

```python
#the number of matches each team won by batting first
a=winbr['winner'].value_counts()
a
```

```python
#top 5 teams won by batting first
b=a.head()
```

```python
#bar plot for the matches won by batting first
plt.figure(figsize=(15,7))
x=list(winbr['winner'].value_counts()[0:5].keys())
y=list(winbr['winner'].value_counts()[0:5])
plt.bar(x,y)
```

```python
#pie chart for wins by batting first
plt.pie(y,labels=x,autopct='%0.2f%%')
plt.show()
```
