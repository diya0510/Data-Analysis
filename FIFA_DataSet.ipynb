**FIFA DataSet Analysis**

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
import numpy as np
```

```python
fifa=pd.read_csv('players_20.csv')
fifa.head()
```

```python
#printing different columns
for col in fifa.columns:
  print(col)
```

```python
fifa.shape
```

```python
#number of players from each country
fifa['nationality'].value_counts()
```

```python
#top ten countries with highest number of players
fifa['nationality'].value_counts()[0:10]
```

```python
#top 5
fifa['nationality'].value_counts().head()
```

```python
plt.figure(figsize=(15,7))
plt.bar(list(fifa['nationality'].value_counts().head().keys()),list(fifa['nationality'].value_counts().head()))
plt.xlabel('Countries')
plt.ylabel('No. of Players')
plt.title("PLAYERS OF DIFFERENT COUNTRIES")
plt.show()
```

```python
#wages of different players
player_salary=fifa[['short_name','wage_eur']]
player_salary
```

```python
#first 5 players w.r.t wages
player_salary.head()
```

```python
#top 5 highest paid players
top5=player_salary.sort_values(by=['wage_eur'],ascending=False).head()
```

```python
#bar plot
plt.figure(figsize=(15,7))
plt.bar(top5['short_name'],top5['wage_eur'],color=['blue','yellow','red','green','purple'])
plt.show()
```

```python
german=fifa[fifa['nationality']=='Germany']
german
```

```python
#tallest German player
max_height=german['height_cm'].max()
german[german['height_cm']==max_height]
```

```python
german[['short_name','height_cm']].sort_values(by=['height_cm'],ascending=False)
```

```python
#sort German players according to their weight
german.sort_values(by=['weight_kg'],ascending=False).head()
```

```python
#top 5 highest paid german players
german[['short_name','wage_eur']].sort_values(by=['wage_eur'],ascending=False).head()
```

```python
#best rating for best shooting among all the players
rankforshoot=fifa[['short_name','shooting']]
rankforshoot
```

```python
#top 5 players according to their shhoting rank
rankforshoot.sort_values(by=['shooting'],ascending=False).head()
```

```python
#rating based on defending skills
rankfordefend=fifa[['short_name','defending']]
rankfordefend
```

```python
rankfordefend.sort_values(by=['defending'],ascending=False).head()
```

```python
realmadrid=fifa[fifa['club']=="Real Madrid"]
realmadrid.head()
```

```python
#highest paid player in the Real Madrid club
realmadrid.sort_values(by=['wage_eur'],ascending=False).head()
```
